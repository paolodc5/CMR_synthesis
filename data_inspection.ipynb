```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# load tissue prop
tissue_prop_path = "./data/ACDC_tissue_prop/patient001_ED.npz"

if os.path.exists(tissue_prop_path):
    tissue_prop = np.load(tissue_prop_path)

print(tissue_prop.files)  # List all arrays stored in the .npz file
```

```python
tissue_prop['image'].shape  # Example of accessing the 'image' array
```

```python
img_to_show = tissue_prop['image'][0,5,...]
```

```python
cols = 3
rows = tissue_prop['image'].shape[1]//cols + 1

fig, axes = plt.subplots(rows, cols, figsize=(12, 12))
idx_slice = np.linspace(0, tissue_prop['image'].shape[1]-1, tissue_prop['image'].shape[1], dtype=int)


for i, ax in enumerate(axes.flatten()):
    
    idx = idx_slice[i]
    img = tissue_prop['image'][0, idx, ...]
    
    ax.imshow(img, cmap='gray')
    ax.set_title(f"Slice {idx}")
    ax.axis('off')  

    if i >= tissue_prop['image'].shape[1] - 1:
        break
plt.tight_layout()
plt.show()
```
